```python
def select_random_chann(events, lenght) :

    new_list = []
    rand_x = np.random.randint(0, 4095, lenght)
    rand_y = np.random.randint(0, 4095, lenght) 
    for e in events:
       # print(e.x, e.y)
        for ch in range(lenght):
            print(rand_x[ch])
            print(e.x)
            if (e.x == rand_x[ch] and e.y == rand_y[ch]):
             #   print(e.x, e.y)
                print(1)
                new_list.append(e)  
                print(new_list)

    return new_list
```
