```python
import scipy.io
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
CM500ms = scipy.io.loadmat('CM_recordings/CM500ms.mat')
CM20s = scipy.io.loadmat('CM_recordings/CM20sec.mat')

CM20s = CM20s["rec1_mV"]
CM500ms=CM500ms["rec1_mV"]

time_CM20S= np.linspace(0,20,20*20000+1)
time_CM500ms= np.linspace(0,500,500*20+1)

```

```python
plt.figure(figsize=(20,5))
plt.plot(time_CM20S, CM20s)
plt.ylabel("Voltage [mV]")
plt.xlabel("Time [s]")
```

```python
plt.figure(figsize=(10,5))
plt.plot(time_CM500ms,CM500ms)
plt.ylabel("Voltage (V)")
plt.xlabel('Time [ms]')
plt.title('Pre recorder CM spike')
plt.xlim([0,300])

```

```python
df = pd.read_csv('1901_BEL_data/scope_87.csv')

time= df["x-axis"][1:-1].to_numpy().astype(float)
CM= df["1"][1:-1].to_numpy().astype(float)
DN= df["3"][1:-1].to_numpy().astype(float)
UP= df["4"][1:-1].to_numpy().astype(float)

```

```python



fig, [ax1,ax2] = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

ax1.plot(time, CM, 'k')
ax1.set_ylim([-1,1])
ax1.set_ylabel("Voltage [mV]")
ax2.plot(time, UP,"r")
ax2.set_ylim([-0.5,0.5])
ax2.set_ylabel("Voltage [V]")
ax2.set_xlabel("Time [ms]")
```
