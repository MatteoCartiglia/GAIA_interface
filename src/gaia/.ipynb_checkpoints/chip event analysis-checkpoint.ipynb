```python
import numpy as np
import os
import pandas as pd
import struct
import matplotlib.pyplot as plt
```

```python
class event2d():
    def __init__(self, ts, x, y, p):
        self.ts = ts   
        self.x = x
        self.y = y
        self.p = p
```

```python
fileName = "Recordings/many_events.dat"
with open(fileName, mode='rb') as file: # b is important -> binary
    fileContent = file.read()
```

```python
code ="QIIII"
unpacked = struct.unpack(code * ((len(fileContent)) // 24), fileContent)
events = []
for i in range(len(unpacked)//5):
    ev = event2d(unpacked[i*len(code)+0],unpacked[i*len(code)+1], unpacked[i*len(code)+2], unpacked[i*len(code)+3])
    events.append(ev)
```
